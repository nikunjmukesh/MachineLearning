```python
mapping = {
    ' Other': 0,
    ' Amer-Indian-Eskimo': 1,
    ' Asian-Pac-Islander': 2,
    ' Black': 3,
    ' White': 4,
}
trainSet['Race'] = trainSet['Race'].map(mapping)
testSet['Race'] = testSet['Race'].map(mapping)

#

testSet.head()
```

```python
# mapping unique values from Native contry column to numbers
tem_df = trainSet['Native Country'].unique().tolist()
tem_dic = dict(zip(tem_df, range(len(tem_df))))
trainSet['Native Country'] = trainSet['Native Country'].map(tem_dic)

# mapping unique values from Native contry column to numbers
tem_df = testSet['Native Country'].unique().tolist()
tem_dic = dict(zip(tem_df, range(len(tem_df))))
testSet['Native Country'] = testSet['Native Country'].map(tem_dic)
```

```python
#drop education
trainSet = trainSet.drop(['Education', 'Capital Gain', 'Capital Loss'], axis =1)
testSet = testSet.drop(['Education', 'Capital Gain', 'Capital Loss'], axis =1)
```

```python
mapping = {
    ' <=50K': 1,
    ' >50K': 0
}
mapping1 = {
    ' <=50K.': 1,
    ' >50K.': 0
}
trainSet['Income'] = trainSet['Income'].map(mapping)
testSet['Income'] = testSet['Income'].map(mapping1)
```

```python
X = trainSet[['Age', 'Work Class', 'Education Number', 'Marital Status', 'Occupation', 'Relationship', 'Race', 'Sex', 'Hours Per Week', 'Native Country']]
Y = trainSet['Income']   
#Y = pd.get_dummies(trainSet['Income']).as_matrix()
Y = Y.reshape([Y.shape[0], 1])
```

```python
X_Test = testSet[['Age', 'Work Class', 'Education Number', 'Marital Status', 'Occupation', 'Relationship', 'Race', 'Sex', 'Hours Per Week', 'Native Country']]
Y_Test = testSet['Income']
#YTest = pd.get_dummies(testSet['Income']).as_matrix()
Y_Test = Y_Test.reshape([Y_Test.shape[0], 1])
```

```python

print(X.shape)
print(Y.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y )
```

```python
lr = 0.3
batch_size = 128
n_epochs = 2500

X_hold = tf.placeholder(tf.float32,[None,X.shape[1]])
Y_hold = tf.placeholder(tf.float32,[None,Y.shape[1]])
```

```python
w = tf.Variable(tf.truncated_normal(shape=[X.shape[1], 20], stddev=0.01), name = "w")
b = tf.Variable(tf.constant(0.1, shape=[1,1]), 'b')
h1 = tf.nn.sigmoid(tf.matmul(X_hold,w)+b)


w1 = tf.Variable(tf.truncated_normal(shape=[20, 10], stddev=0.01), name = "w")
b1 = tf.Variable(tf.constant(0.1, shape=[1,1]), 'b')
h2 = tf.nn.sigmoid(tf.matmul(h1,w1)+b1)

w2 = tf.Variable(tf.truncated_normal(shape=[10, Y.shape[1]], stddev=0.01), name = "w")
b2 = tf.Variable(tf.constant(0.1, shape=[1,1]), 'b')


logits = tf.matmul(h2,w2)+b2
sigm = tf.nn.sigmoid(logits, name = 'pred')
pred = tf.cast(tf.greater_equal(sigm, 0.5), tf.float32, name = 'pred')
loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=logits, labels = Y_hold))
opt = tf.train.GradientDescentOptimizer(learning_rate=lr).minimize(loss)
```

```python
cp = tf.equal(pred, Y_hold)
acc = tf.reduce_mean(tf.cast(cp,tf.float32))
sess = tf.InteractiveSession()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
for i in range(29000):
    sess.run(opt, feed_dict={X_hold: X_train, Y_hold: y_train})
    if (i % 100 == 0):
        l = loss.eval(feed_dict={X_hold: X_train, Y_hold: y_train})
        a = acc.eval(feed_dict={X_hold: X_train, Y_hold: y_train})
        print ("loss: {}, accuracy: {}".format(l, a))
a = acc.eval(feed_dict={X_hold: X_test, Y_hold: y_test})
print(a)
```
